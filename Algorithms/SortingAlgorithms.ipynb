```python
#This python script demonstrates what and how to use a Bubble Sort
#
#   Bubble Sort:     One of the simplist algorithems and it works by repeatedly swapping adjacent elements if they are in the wrong order until the entire 
#                    list is sorted
#           

#
#   Key Terminology:
#           Comparision Function: A function that defines the order of the elements during sorting. 
#                                 It takes two elements as an input and returns a value indicating the relative order of elements
#
#           Stable Sorting:       A sorting algorithm is stable if it maintains the relative ortder of elements with equal keys (meaning values that equal eachother are sorted correctly, i.e people with the same first name)
#
#           In-Place Sorting:     Rearranges elements within the input array without requiring additional memory 
#
#           Time Complexity:      Amount of time taken by an algorithm to run
#
#           Space Complexity:     Amount of additional memory used by an algorithem 
#           
#   Steps:
#               1. Iterate through the List:    Start by iterating through the list from the beginning. Compare each element with the element next to it.
#
#               2. Compare and Swap:            Compare the current element with the next element. If the current element is greater (or smaller, depending on whether you're sorting in ascending or descending order) than the next element, swap them. 
#                                               This step ensures that the largest (or smallest) element "bubbles up" to the end of the list.
#
#               3. Repeat the Process:          Continue iterating through the list, comparing and swapping adjacent elements, until no more swaps are needed during a full pass through the list. 
#                                               This means that the largest (or smallest) element has reached its final sorted position. 
#
#               4. Repeat Until Sorted:         Repeat steps 1-3 for the entire list until no more swaps are needed during a full pass through the list. At this point, the entire list is sorted.
#  

import random


unsorted_values_bubble = []

for x in range(0,10):
    unsorted_values_bubble.append(random.randint(1,100))



def bubble_sort(array):
    index = len(array)

    for x in range(index):

        for y in range(0, index-x-1):
        
            if array[y] > array [y+1]:
                array[y], array[y+1] = array[y+1], array[y]
                
       

    return array



print(f"Unsorted Values: {unsorted_values_bubble}  \n")

sorted_values = bubble_sort(unsorted_values_bubble)

print(f"\nSorted Values: {sorted_values}  \n")
```

```python
# This python script demonstrates what and how to use a Selection Sort
#
#   Selection Sort:   An In-Place comparison sorting algorithm. It works by repeatedly finding the minimum element from the unsorted portion of the list and placing it at the beginning.
#                     It uses two subarrays: the first being the sorted subarray at the beginning and the second being the unsorted subarray at the end
#        
#   Steps:
#               1. Find the Minimum:                Start by considering the first element of the list as the minimum (or maximum, depending on whether you're sorting in ascending or descending order). This element is treated as the first element in the sorted part.
#
#               2. Iterate through Unsorted Part:   Begin iterating through the unsorted part of the list, starting from the second element.
#
#               3. Find the Actual Minimum:         While iterating through the unsorted part, compare each element with the current minimum. If you find an element smaller (or larger, depending on the order) than the current minimum, update the minimum to be that element. 
#
#               4. Swap and Repeat:                 After completing the iteration through the unsorted part, swap the found minimum element with the first element in the unsorted part. This effectively moves the minimum element to the end of the sorted part. 
#                                                   Then, repeat steps 2-4 for the remaining unsorted part of the list

unsorted_values_selection = []

for x in range(0,10):
    unsorted_values_selection.append(random.randint(1,100))


def selection_sort(array):
    index = len(array)

    for x in range(index):
        min_element = x

        for y in range(x+1, index):
            if array[y] < array[min_element]:
                min_element = y
        
        #Swap the min element with the first element of the unsorted part
        array[x], array[min_element] = array[min_element], array[x]
        
    
    return array


print(f"Unsorted Values: {unsorted_values_selection}  \n")

sorted_values = selection_sort(unsorted_values_selection)

print(f"\nSorted Values: {sorted_values}  \n")
```

```python
#This python script demonstrates what and how to use a Insertion Sort
#
#   Insertion Sort:   Another simple sort based on comparisions. It works by dividing the list into sorted and unsorted parts. 
#                     It iterates through the unsorted part, sleectin one element at a time, and inserts it into the correct position within the sorted part
#        
#   Steps:
#               1. Divide the List:                 Start by considering the first element of the list as a sorted part and the rest of the elements as an unsorted part. 
#
#               2. Iterate through Unsorted Part:   Begin iterating through the unsorted part of the list, one element at a time, typically starting from the second element.
#
#               3. Insert Elements:                 For each element in the unsorted part, compare it with the elements in the sorted part and find the correct position for insertion. Shift elements in the sorted part to make room for the new element.
#
#               4. Repeat:                          Continue this process until you've iterated through all elements in the unsorted part, effectively expanding the sorted section until the entire list is sorted.
#


unsorted_values_insertion = []

#filling an array with test data
for x in range(0,10):
    unsorted_values_insertion.append(random.randint(1,100))

def insertion_sort(array):
    n = len(array)

    for x in range(1, n):
        key = array[x]
        j = x-1

        while j >= 0 and array[j] > key:
            array[j+1] = array[j]
            j -= 1
        
        array[j+1] = key

    return array

print(f"Unsorted Values: {unsorted_values_insertion}  \n")

sorted_values = insertion_sort(unsorted_values_insertion)

print(f"\nSorted Values: {sorted_values}  \n")
```

```python
#This python script demonstrates what and how to use a Merge Sort
#
#   Insertion Sort:   Another simple sort based on comparisions. It works by dividing the list into sorted and unsorted parts. 
#                     It iterates through the unsorted part, sleectin one element at a time, and inserts it into the correct position within the sorted part
#        
#   Steps:
#               1. Divide the List:                 Divide the unsorted list into two halves
#
#               2. Recursively Sort:                 Recursively sort each half by applying Merge Sort
#
#               3. Merge:                           Merge the two sorted halves back together to obtain the final sorted list
#

unsorted_values_merge= []

#filling an array with test data
for x in range(0,10):
    unsorted_values_merge.append(random.randint(1,100))


def merge_sort(array):
    
    if len (array) > 1 :
        #find middle of the array
        middle = len(array)//2

        #divide arrays by left and right
        left = array[:middle]
        right = array[middle:]

        #recursive calls for the 2 halves
        merge_sort(left)
        merge_sort(right)

        #merge both halves
        i = j = k = 0

        while i < len(left) and j < len(right):

            if(left[i] < right[j]):
                array[k] = left[i]
                i+=1
            else:
                array[k] = right[j]
                j+=1
            k+=1
            
        while i < len(left):
            array[k] = left[i]

            i += 1
            k += 1

        while j < len(right):
            array[k] = right[j]
            j += 1
            k += 1
        
        return array

print(f"Unsorted Values: {unsorted_values_merge}  \n")

sorted_values = merge_sort(unsorted_values_merge)

print(f"\nSorted Values: {sorted_values}  \n")
```

```python
#This python script demonstrates what and how to use a Quick Sort
#
#   Quick Sort:       More widely used sorting algorithm known for its efficiency and low memory requirements. it follows the divide-and-conquer appreach of 
#                     partitioning the list inot two sublists based on a pivot element(basically the middle of the list), and then recursively sorts the sublists
#                     
#        
#   Steps:
#               1. Choose the pivot element     Select a pivot element from the array. The choice of the pivot can significantly affect the efficiency of the algorithm. 
#                                               Common strategies include choosing the first element, the last element, the middle element, or a randomly selected element as the pivot.
#
#               2. Partitioning                 Rearrange the elements in the array such that all elements less than the pivot are on the left, and all elements greater than the pivot are on the right. 
#                                               This step effectively divides the array into two subarrays, one with elements less than the pivot and the other with elements greater than the pivot. 
#                                               The pivot itself is now in its final sorted position.
#
#               3. Recursion                    Recursively apply the Quick Sort algorithm to the two subarrays created in step 2. Continue this process until the entire array is sorted. 
#                                               The base case for recursion is when a subarray has only one or zero elements, as it is already considered sorted.
#
#               4. Combine Subarrays            Combine the sorted subarrays from step 3 to produce the final sorted array. The combined result will have all elements sorted in the correct order.


unsorted_values_quick = []

#filling an array with test data
for x in range(0,10):
    unsorted_values_quick.append(random.randint(1,100))

def quick_sort(array):

    if len(array) <= 1:
        return array
    
    #finds the middle element
    pivot = array[len(array) // 2]
    
    #Each of these arguments creates a new list with certain criteria, if the value is less than pivot than it goes to the left list, if the value is equal to the pivot then middle list, if the value is greater than the pivot then right list
    left = [x for x in array if x < pivot]
    middle = [x for x in array if x == pivot]
    right = [x for x in array if x > pivot]

    return quick_sort(left) + middle + quick_sort(right)


print(f"Unsorted Values: {unsorted_values_quick}  \n")

sorted_values = merge_sort(unsorted_values_quick)

print(f"\nSorted Values: {sorted_values}  \n")
```
