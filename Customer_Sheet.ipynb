```python
import pandas as pd
import openpyxl
```

```python
df = pd.read_excel(r"C:\Users\Griffin Bohannon\Desktop\Coding\Python\Projects\Customer_Analysis\Customer Call List.xlsx")
df
```

```python
df = df.drop_duplicates()
```

```python
df.drop(columns = "Not_Useful_Column")
```

```python
#Used the strip function to get rip of random values and symbols (mainly to get practice with the strip funciton)
df["Last_Name"] = df["Last_Name"].str.lstrip("...")
df["Last_Name"] = df["Last_Name"].str.lstrip("/")
df["Last_Name"] = df["Last_Name"].str.rstrip("_")
df

#Could use the strip function like this 
#   df["Last_Name"] = df["Last_Name"].str.strip("123._/") 
```

```python
#Cleaning data of everything other than letters and numbers ('^' means everything except) to make everything similar
df["Phone_Number"] = df["Phone_Number"].str.replace('[^0-9]','',regex=True)

#Using lambda to clean and format the data to the proper phone number format

df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))

df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])

#Removing all other random null values, etc.
df["Phone_Number"] = df["Phone_Number"].str.replace('nan--','')
df["Phone_Number"] = df["Phone_Number"].str.replace('--','')
df
```

```python
#Spliting the Address Column on the commas to get Address, State, and ZipCode
df[["Address", "State", "Zip_Code"]] = df["Address"].str.split(',', expand=True)
df
```

```python
#Using the Replace Function to get some uniformity with the code 
df["Paying Customer"] = df["Paying Customer"].str.replace('Yes', 'Y')
df["Paying Customer"] = df["Paying Customer"].str.replace('No', 'N')

df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes', 'Y')
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No', 'N')
df
```

```python
#Getting ride of 'N/A's and Null Values(NaN)
df = df.replace('N/A','')
df = df.fillna('')
df
```

```python
#Removing Records of Customers who do not want to be contacted

for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y':
        df.drop(x, inplace=True)

df
```

```python
#Removing Records of Customers that do not have a phone number inputed
for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
        df.drop(x, inplace=True)

df
```

```python
#Reseting the Index for better readability 
df = df.reset_index(drop=True)
```
